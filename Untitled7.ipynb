```python
#data filter
dfa = df1[df1['Restaurants that have installed device'] == 1]
```

```python
#group means
dfa.groupby('Dates when device is installed')['Net Sales', 'Game Revenue (formatted)', 'Labor Cost', 'POS Checks'].mean()
```

```python
#data filter
dfc1 = dfa[dfa['Dates when device is installed'] == 0]
dfc2 = dfa[dfa['Dates when device is installed'] == 1]
```

```python
# F test - test for equality of variance
from scipy import stats
stats.levene(dfc1['Net Sales'], dfc2['Net Sales'])
```

```python
#two tail Welch's t test
stats.ttest_ind(dfc1['Net Sales'], dfc2['Net Sales'], equal_var=False )
```

```python
# one tail Welch's t test
results =  stats.ttest_ind(dfc1['Net Sales'], dfc2['Net Sales'], equal_var=False)
alpha = 0.05
if (results[0] > 0) & (results[1]/2 < alpha):
    print ("reject null hypothesis")
else:
    print ("do not reject null hypothesis")
```

```python
# F test - test for equality of variance
stats.levene(dfc1['Labor Cost'], dfc2['Labor Cost'])
```

```python
#two tail independent t test
stats.ttest_ind(dfc1['Labor Cost'], dfc2['Labor Cost'])
```

```python
# one tail independent t test
results2 =  stats.ttest_ind(dfc1['Labor Cost'], dfc2['Labor Cost'])
alpha = 0.05
if (results2[0] > 0) & (results2[1]/2 < alpha):
    print ("reject null hypothesis")
else:
    print ("do not reject null hypothesis")
```

```python
#Predictive Model
#data scaling to put numeric variables on the same scale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[['Net Sales','Labor Cost','POS Checks']] = scaler.fit_transform(df1[['Net Sales','Labor Cost','POS Checks']])
```

```python
import statsmodels.api as sm # import statsmodels
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor

#predictor variables
X = df1[['Net Sales','Labor Cost','POS Checks','Dates when device is installed']]

#check VIF
pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns)
```

```python

#predictor variables
X = df1[['Net Sales','Labor Cost','Dates when device is installed']]

#check VIF
pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns)
```

```python
X = sm.add_constant(X)
#add constant
```

```python
#data split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    df1['Game Revenue (formatted)'], 
                                                    test_size=0.3,random_state=101)
```

```python

# Model fitting
results = sm.OLS(y_train, X_train).fit() ## sm.OLS(output, input)
#Model prediction
predictions = results.predict(X_test)
```

```python
# Model statistics
results.summary()
```

```python
#Breusch pagan test for homogenity
import statsmodels

name = ['Lagrange multiplier statistic', 'p-value', 
        'f-value', 'f p-value']
bp = statsmodels.stats.diagnostic.het_breushpagan(results.resid, results.model.exog)
bp
pd.DataFrame(name,bp)
```

```python
#normality test
from scipy import stats
stats.kstest(results.resid, 'norm')
```

```python

import statsmodels.formula.api as smf

# Model fitting
results2 = smf.OLS(y_train, X_train).fit(cov_type='HC0') ## sm.OLS(output, input)
#Model prediction
predictions = results2.predict(X_test)

# Model statistics
results2.summary()
```
